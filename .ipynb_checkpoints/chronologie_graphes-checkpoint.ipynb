```python
from packages import *
from functions import *
import datetime as dt
import seaborn as sns
```

```python
df = pd.read_csv('data/data_concat_clean/base_complete_022023.csv')
```

```python
# Convertir la colonne 'created_at' en datetime et créer une nouvelle colonne 'month_year'
df['created_at'] = pd.to_datetime(df['created_at'])
df['month_year'] = df['created_at'].dt.to_period('M')

# Filtrer les données pour ne garder que les tweets après 2016
df = df[df['created_at'] >= '2017-01-01']

# Compter le nombre de tweets par mois pour les sources 'api' et 'ina'
pivot_counts = pd.pivot_table(df[df['source_base'].isin(['api', 'ina'])], 
                              values='tweet_id', 
                              index=['month_year'], 
                              columns=['source_base'], 
                              aggfunc='count',
                              fill_value=0)
### Liste des dates clés
#octobre 2017: metoo et btp
#janvier 2018 : tribune liberté d'importuner
#octobre 2018 : omerta EN
#fevrier 2020 : Cesar, polanski & haenel
#juin 2020 : #iwas
#janvier 2021: lancement du MeTooInceste, MeTooGay, Familia Grande
#mai 2022 : mediapart et affaire ppda

dates_clefs = ['2017-10', '2018-01', '2018-10', '2020-06', '2021-01', '2022-05' ]
labels = ['', "Tribune Le Monde", "#pasdevague", '#iwas', "#metooinceste & #metoogay", "Affaire PPDA dans Mediapart"]


# Créer un plot avec les options souhaitées
fig, ax = plt.subplots(figsize=(15, 6))
pivot_counts.plot(ax=ax, kind='line', linewidth=1, alpha=0.5, linestyle='-', color=['blue', 'red'])
ax.set_xlabel('')
ax.set_ylabel('Nombre de tweets publiés')
ax.set_title('')

# Ajouter des traits verticaux à certaines dates clés
#for date in dates_clefs:
 #   ax.axvline(x=date, color='green', linestyle='--', linewidth=1)
  #  ax.annotate(date, xy=(date, 1), xycoords=('data', 'axes fraction'), xytext=(0, 10), 
   #             textcoords='offset points', ha='center', arrowprops=dict(arrowstyle="->", color='gray'))
    
for date, label in zip(dates_clefs, labels):
    ax.axvline(x=date, color='green', linestyle='--', linewidth=1)
    ax.annotate(label, xy=(date, 1), xycoords=('data', 'axes fraction'), xytext=(-10, 10), 
                textcoords="offset pixels", ha='left', arrowprops=dict(arrowstyle="->", color='gray'))
    
#ajouter le 8 mars
dates_mars = [f'{annee}-03-08' for annee in range(2017, 2023)]

#for date in dates_mars:
#    ax.axvline(x=date, color='gray', linestyle='--', linewidth=1)
    
# Ajouter la grille
ax.grid(True)


# Colorer le fond du plot par année
for i, year in enumerate(range(2017, 2023)):
    ax.axvspan(dt.datetime(year, 1, 1), dt.datetime(year+1, 1, 1), 
               facecolor=f'C{i%2}', alpha=0.1)

# Afficher le plot
plt.tight_layout()
plt.legend()
plt.show()
```

```python
df['metoo'] = df['text'].str.contains('#metoo', case=False)
df['metooinceste'] = df['text'].str.contains('#metooinceste', case=False)
df['balancetonporc'] = df['text'].str.contains('#balancetonporc', case=False)
df['metoogay'] = df['text'].str.contains('#metoogay', case=False)
df['iwas'] = df['text'].str.contains('#iwas', case=False)
# Convertir la colonne created_at en datetime et extraire le mois et l'année
df['created_at'] = pd.to_datetime(df['created_at'])
df['month_year'] = df['created_at'].dt.to_period('M')

# Regrouper les données par mois et par hashtag et compter le nombre de tweets pour chaque groupe
grouped = df.groupby('month_year')[['metoo', 'metooinceste', 'balancetonporc', 'metoogay', 'iwas']].sum()

### Liste des dates clés
#octobre 2017: metoo et btp
#janvier 2018 : tribune liberté d'importuner
#octobre 2018 : omerta EN
#fevrier 2020 : Cesar, polanski & haenel
#juin 2020 : #iwas
#janvier 2021: lancement du MeTooInceste, MeTooGay, Familia Grande
#mai 2022 : mediapart et affaire ppda

dates_clefs = ['2017-10', '2018-01', '2018-10', '2020-06', '2021-01', '2022-05' ]
labels = ['', "Tribune Le Monde", "#pasdevague", '#iwas', "#metooinceste & #metoogay", "Affaire PPDA dans Mediapart"]


# Créer un plot avec les options souhaitées
fig, ax = plt.subplots(figsize=(15, 6))
grouped.plot(ax=ax, kind='line', linewidth=1, alpha=1, linestyle='-')
ax.set_xlabel('')
ax.set_ylabel('Nombre de tweets publiés')
ax.set_title('')

# Ajouter des traits verticaux à certaines dates clés
#for date in dates_clefs:
 #   ax.axvline(x=date, color='green', linestyle='--', linewidth=1)
  #  ax.annotate(date, xy=(date, 1), xycoords=('data', 'axes fraction'), xytext=(0, 10), 
   #             textcoords='offset points', ha='center', arrowprops=dict(arrowstyle="->", color='gray'))
    
for date, label in zip(dates_clefs, labels):
    ax.axvline(x=date, color='gray', linestyle='--', linewidth=1, alpha=0.7)
    ax.annotate(label, xy=(date, 1), xycoords=('data', 'axes fraction'), xytext=(-10, 10), 
                textcoords="offset pixels", ha='left', arrowprops=dict(arrowstyle="->", color='gray'))
    
#ajouter le 8 mars
dates_mars = [f'{annee}-03-08' for annee in range(2017, 2023)]

#for date in dates_mars:
#    ax.axvline(x=date, color='gray', linestyle='--', linewidth=1)
    
# Ajouter la grille
ax.grid(True)


# Colorer le fond du plot par année
for i, year in enumerate(range(2017, 2023)):
    ax.axvspan(dt.datetime(year, 1, 1), dt.datetime(year+1, 1, 1), 
               facecolor=f'C{i%2}', alpha=0.1)

# Afficher le plot
plt.tight_layout()
plt.legend()
plt.show()
```
