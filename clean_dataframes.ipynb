### base INA

```python
metoo_ina = pd.read_csv('data\\ina\\es_dump_twitter_metoo.excel.csv', sep= ';')
inceste_ina = pd.read_csv('data\\ina\\es_dump_twitter_metooinceste.excel.csv', sep= ';')
gay_ina = pd.read_csv('data\\ina\\es_dump_twitter_metoogay.excel.csv', sep= ';')
iwas_ina = pd.read_csv('data\\ina\\es_dump_twitter_iwas.excel.csv', sep= ';')
btp_ina = pd.read_csv('data\\ina\\es_dump_twitter_balancetonporc.excel.csv', sep= ';')
```

```python
metoo_ina_clean = clean_hashtags(metoo_ina,hashtags = "metoo|Metoo|me too|Me Too")
metoo_ina_clean = remove_rt(metoo_ina_clean)
metoo_ina_clean.text = metoo_ina_clean.text.map(lambda x: re.sub('[,\.!?"()]', '', x)).apply(clean_text)

inceste_ina_clean = clean_hashtags(inceste_ina,hashtags = "metooinceste|Metooinceste|me too inceste|Me Too inceste")
inceste_ina_clean = remove_rt(inceste_ina_clean)
inceste_ina_clean.text = inceste_ina_clean.text.map(lambda x: re.sub('[,\.!?"()]', '', x)).apply(clean_text)

gay_ina_clean = clean_hashtags(gay_ina,hashtags = "metoogay|Metoogay|me too gay|Me Too gay")
gay_ina_clean = remove_rt(gay_ina_clean)
gay_ina_clean.text = gay_ina_clean.text.map(lambda x: re.sub('[,\.!?"()]', '', x)).apply(clean_text)

iwas_ina_clean = clean_hashtags(iwas_ina,hashtags = "iwas")
iwas_ina_clean = remove_rt(iwas_ina_clean)
iwas_ina_clean.text = iwas_ina_clean.text.map(lambda x: re.sub('[,\.!?"()]', '', x)).apply(clean_text)

btp_ina_clean = clean_hashtags(btp_ina,hashtags = "btp|BalanceTonPorc|Balance ton porc|balancetonporc|Balance Ton Porc")
btp_ina_clean = remove_rt(btp_ina_clean)
btp_ina_clean.text = btp_ina_clean.text.map(lambda x: re.sub('[,\.!?"()]', '', x)).apply(clean_text)
```

```python
ina_concat = pd.concat([btp_ina_clean, iwas_ina_clean, gay_ina_clean, inceste_ina_clean, metoo_ina_clean]).drop_duplicates('id')
```

```python
ina_concat.columns = (['tweet_id','created_at','author_id','author_username','author_name', 'author_created_at',
                       'author_followers_count','author_tweet_count','text','hashtags','mentions','user_mentions.screen_name',
                      'user_mentions.name','emojis', 'emojis_utf8', 'lang','quote_status','quoted_status.id', 'quoted_status.user.id',
                       'quoted_status.user.screen_name'])
```

```python
ina_concat.to_csv('base_ina_concat_clean.csv')
```
